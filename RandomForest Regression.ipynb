```python
import graphlab as gl
```

```python
data = gl.SFrame.read_csv('http://s3.amazonaws.com/gl-testdata/xgboost/mushroom.csv')
```

```python
data
```

```python
data.show()
```

```python
data['label'] = data['label'] =='p'
```

```python
data
```

```python
train,test = data.random_split(.8,seed = 0)
```

```python
train
```

```python
model = gl.random_forest_regression.create(train,target='label')
```

```python
model
```

```python
predictions = model.predict(test)
```

```python
predictions
```

```python
results = model.evaluate(test)
```
